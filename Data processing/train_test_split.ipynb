```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("./data_5_freq.csv")

df_blank = df.loc[(df["1"] == False) & (df["2"] == False) & (df["3"] == False) & (df["4"] == False) & (df["5"] == False) & (df["6"] == False) & (df["7"] == False) & (df["8"] == False)]
```

```python
len(df)
```

```python
df_no_blank = df.loc[~df["id"].isin(df_blank["id"])]
```

```python
X = df_no_blank[["id", "text"]]
y = df_no_blank[["id", "1", "2", "3", "4", "5", "6", "7", "8"]]
```

```python
train, test, _, _ = train_test_split(df_no_blank, df_no_blank, test_size=0.11, random_state=42)
```

```python
new_train = pd.concat([train, df_blank])
new_train
```

```python
train_len = len(new_train)
test_len = len(test)

for i in range(1, 9):
    num_train_true = new_train[str(i)].sum()
    num_test_true = test[str(i)].sum()
    
    print(f"Tag {i}")
    print(f"True tag pct train: {float(num_train_true) / float(train_len) * 100}")
    print(f"True tag pct test: {float(num_test_true) / float(test_len) * 100}")
```

```python
id_test = list(test["id"].values)
```
