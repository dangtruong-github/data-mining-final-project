```python
import numpy as np
import pandas as pd
import json
```

```python
filename = "./dictionary_after_cleaned_by_code.txt"

word_dict = {}

with open(filename, encoding = 'utf-8') as f:
    for x in f:
        splitting = x.split(": ")
        splitting[1] = int(splitting[1][:-1])
        print(splitting)
        
        word_dict[splitting[0]] = splitting[1]
```

```python
word_dict_10 = {key: item for key, item in word_dict.items() if item >= 10}
```

```python
len(word_dict_10.keys())
```

```python
word_dict_5 = {key: item for key, item in word_dict.items() if item >= 5}
```

```python
len(word_dict_5.keys())
```

```python
with open("./word_dict_10.json", "w") as outfile:
    json.dump(word_dict_10, outfile)
```

```python
with open("./word_dict_5.json", "w") as outfile:
    json.dump(word_dict_5, outfile)
```
