```python
df = pd.read_csv("./6_filter_latin_tags.csv")

tags_hash_table = {
    1: ["romance", "love", "langman", "ngot", "sung", "ngontinh", "tinhcam", "lovestory", "teenromance", "loveatfirstsight", 
        "chicklit", "heartbreak", "yeu", "thoca"],
    2: ["drama", "kinhdi", "horror", "nguoisoi", "macarong", "wolves", 'werewolf', 'paranormal', 'werewolves', 'pack', 
        "adventure", "phieuluu", "bian", "mystery", "secrets", "mysterious", "action", "hanhdong", "thriller", "trinhtham", 
        "killer", "phaan", "doan"],
    3: ["fanfic", "fanfiction", "fandom", "teenfic", "teenfiction", "fiction", "fairytail", "fic", "hucau", 
        "fantasy", "vientuong", "supernatural", "magic", "truyentamlinh", "sieunhien"],
    4: ["lgbt", "lesbian", "dammy", "bl", "bachhop", "boyxboy", "gay", "yaoi", "boylove", "girllove", "freethelgbt", "bl", 
        "gl", "songtinh", "kpop", "bts", 'taehyung', 'vkook', 'jungkook'],
    5: ["haihuoc", "humor", "comedy", "funny"],
    6: ["hiendai", "modern", "school", "hocduong", "highschool", "teen", "youngadult"],
    7: ["codai", "classics"],
    8: ["tieuthuyet", "tieuthuyetlichsu", "tieuthuyetthieunien", "novel", "tieuthuyetchung"],
}

tags_hash_key = {}
```

```python
# key value swap
def key_value_swap():
    for key, value in tags_hash_table.items():
        for tag in value:
            tags_hash_key[tag] = key
```

```python
# tags: str to list
def str_to_list(string):
    return string[2:-2].split("', '")

def tags_str_to_list():
    for i in range(len(df)):
        tags = df.iloc[i]["story_tags_list"]

        if isinstance(tags, list):
            continue

        df.at[i, "story_tags_list"] = str_to_list(tags)
```

```python
def tags_string_to_num(printing=True):
    full_num = 0
    
    key_value_swap()
    tags_str_to_list()

    for i in range(len(df)):
        tags = df.iloc[i]["story_tags_list"]
        
        full_num_bool = False

        for index, tag in enumerate(tags):
            if isinstance(tag, int):
                full_num_bool = True
                break
                
            if tag == "fanfic":
                print("here1", end=" ")

            if tag in tags_hash_key.keys():
                if tag == "fanfic":
                    print("here2")
                full_num_bool = True
                break
        
        if full_num_bool:
            new_tags_list = []
            for index, tag in enumerate(tags):
                if isinstance(tag, int):
                    new_tags_list.append(tag)
                    continue
              
                if tag in tags_hash_key.keys():
                    new_tags_list.append(tags_hash_key[tag])
                    continue
                
            full_num += 1       
                
            df.at[i, "story_tags_list"] = new_tags_list

    print(full_num)
    print(len(df))
    
tags_string_to_num()
```

```python
def all_unique_tags():
    #tags_string_to_num(printing=False)
    
    unique_tags = []
    
    for i in range(len(df)):
        tags = df.iloc[i]["story_tags_list"]
        
        unique_tags.extend(tags)
        
    #print(len(Counter(unique_tags).keys()))
    #print(Counter(unique_tags))
    
    tags_with_freq = {key: value for key, value in Counter(unique_tags).items() if isinstance(key, str) and value >= 50}
    
    print(len(Counter(unique_tags)))
    print(Counter(unique_tags))
    print(len(tags_with_freq))
    print(tags_with_freq)
```
