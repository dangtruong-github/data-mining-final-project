```python
def time_to_minutes(time_str):
    to_split = time_str[:-1].split('h ')
    #print(int(to_split[0]))
    #print(int(to_split[1]))
    
    if len(to_split) < 2:
        if time_str[-1] == 'm':
            return int(to_split[0])
        elif time_str[-1] == 'h':
            return int(to_split[0]) * 60
        else:
            return 5
    
    total = int(to_split[0]) * 60 + int(to_split[1])
    
    return total

time1 = "20h 30m"
time2 = "15h 5m"

print("Time 1: {}".format(time_to_minutes(time1)))
print("Time 2: {}".format(time_to_minutes(time2)))
print("Time df: {}".format(time_to_minutes(df2.iloc[90]["Thá»i gian"][10:])))
```

# Clean duplicate data

## Check if same title but different link

```python
for i in range(len(df)):
    break
    for j in range(i + 1, len(df)):
        if i == j:
            continue
    
        if df.at[i, 'title'] == df.at[j, 'title']:
            if df.at[i, 'link'] != df.at[j, 'link']:
                print("Title: {}".format(df.at[i, 'title']))
                print("Link i: {}".format(df.at[i, 'link']))
                print("Link j: {}".format(df.at[j, 'link']))
```

--> Two different stories can have same exact title

```python
clean_df = pd.DataFrame(columns=df.columns)

for i in range(len(df)):
    cur_row = clean_df.loc[(clean_df["link"] == df.iloc[i]["link"])]
    if cur_row.empty:
        clean_df = pd.concat([clean_df, df.iloc[[i]]], ignore_index=True).reset_index().drop(["index"], axis=1)
    else:
        indexes = cur_row.index.values[0]
        cur_row_df = df.iloc[i]
        
        # link
        if (cur_row["title"] != cur_row_df["title"]).values[0]:
            print("Different tite: {}".format(cur_row_df["title"]))
            print("Orig title: {}".format(cur_row["title"].values[0]))
            
            if isinstance(cur_row["title"].values[0], str) == False:
                clean_df.at[indexes, "title"] = cur_row_df["title"]
            elif isinstance(cur_row_df["title"], str):
                break
          
        # depth
        if (cur_row["depth"] != cur_row_df["depth"]).values[0]:
            print("Different depth: {}".format(cur_row_df["title"]))
            print("Orig depth: {}, new depth: {}".format(cur_row["depth"].values[0], cur_row_df["depth"]))
            
            clean_df.at[indexes, "depth"] = min(cur_row["depth"].values[0], cur_row_df["depth"])
        
        # story tags list
        new_row = cur_row_df["story_tags_list"]
        
        if isinstance(new_row, list):
            tags = cur_row["story_tags_list"].values[0]
            if isinstance(tags, list) == False:
                clean_df.at[indexes, "story_tags_list"] = new_row
            else:
                tags.extend(new_row)
                tags = list(set(tags))
                clean_df.at[indexes, "story_tags_list"] = tags
        
        # reading number
        if (cur_row["LÆ°á»£t Ä‘á»c"] != cur_row_df["LÆ°á»£t Ä‘á»c"]).values[0]:
            print("Different LÆ°á»£t Ä‘á»c: {}".format(cur_row_df["title"]))
            print("Orig LÆ°á»£t Ä‘á»c: {}, new LÆ°á»£t Ä‘á»c: {}".format(cur_row["LÆ°á»£t Ä‘á»c"].values[0], cur_row_df["LÆ°á»£t Ä‘á»c"]))
            
            clean_df.at[indexes, "LÆ°á»£t Ä‘á»c"] = max(cur_row["LÆ°á»£t Ä‘á»c"].values[0], cur_row_df["LÆ°á»£t Ä‘á»c"])
        
        # votes number
        if (cur_row["LÆ°á»£t bÃ¬nh chá»n"] != cur_row_df["LÆ°á»£t bÃ¬nh chá»n"]).values[0]:
            print("Different LÆ°á»£t bÃ¬nh chá»n: {}".format(cur_row_df["title"]))
            print("Orig LÆ°á»£t bÃ¬nh chá»n: {}, new LÆ°á»£t bÃ¬nh chá»n: {}".format(cur_row["LÆ°á»£t bÃ¬nh chá»n"].values[0], cur_row_df["LÆ°á»£t bÃ¬nh chá»n"]))
                
            clean_df.at[indexes, "LÆ°á»£t bÃ¬nh chá»n"] = max(cur_row["LÆ°á»£t bÃ¬nh chá»n"].values[0], cur_row_df["LÆ°á»£t bÃ¬nh chá»n"])
        
        # lengths number
        if (cur_row["ChÆ°Æ¡ng"] != cur_row_df["ChÆ°Æ¡ng"]).values[0]:
            print("Different ChÆ°Æ¡ng: {}".format(cur_row_df["title"]))
            print("Orig ChÆ°Æ¡ng: {}, new ChÆ°Æ¡ng: {}".format(cur_row["ChÆ°Æ¡ng"].values[0], cur_row_df["ChÆ°Æ¡ng"]))
                
            clean_df.at[indexes, "ChÆ°Æ¡ng"] = max(cur_row["ChÆ°Æ¡ng"].values[0], cur_row_df["ChÆ°Æ¡ng"])
        
        # recommended
        new_row = cur_row_df["recommended_stories"]
        
        if isinstance(new_row, list):
            recommended = cur_row["recommended_stories"].values[0]
            if isinstance(recommended, list) == False:
                clean_df.at[indexes, "recommended_stories"] = new_row
            else:
                recommended.extend(new_row)
                recommended = list(set(recommended))
                clean_df.at[indexes, "recommended_stories"] = recommended
        
        # badges
        new_row = cur_row_df["badges"]
        
        if isinstance(new_row, list):
            badges = cur_row["badges"].values[0]
            if isinstance(badges, list) == False:
                clean_df.at[indexes, "badges"] = new_row
            else:
                badges.extend(new_row)
                badges = list(set(badges))
                clean_df.at[indexes, "badges"] = badges
        
        # time
        time = cur_row["Thá»i gian"].values[0]
        new_row = cur_row_df["Thá»i gian"]
        if type(time) == type(math.nan):
            clean_df.at[indexes, "Thá»i gian"] = new_row
        elif type(new_row) != type(math.nan):
            if time != new_row:
                print("Different Thá»i gian: {}".format(cur_row_df["title"]))
                print(f"Orig Thá»i gian: {time}, new Thá»i gian: {new_row}")
                
                clean_df.at[indexes, "Thá»i gian"] = max(time, new_row)

clean_df
```

```python
def str_to_list(string):
    return string[2:-2].split("', '")
```

```python
cur_id = 2975
new_story = []

prefix = "https://www.wattpad.com/story/"

for i in range(len(plain_data)):
    recommended_links = plain_data.at[i, "recommended_stories"]
    recommended_stories = []
    
    if type(recommended_links) == type(math.nan):
        continue
        
    recommended_links = str_to_list(recommended_links)
    
    for link in recommended_links:
        if isinstance(link, str) == False:
            continue
            
        if link[:len(prefix)] != prefix:
            continue
            
        story_finder = id_with_link.loc[id_with_link["link"] == link]
        
        if story_finder.empty:
            recommended_stories.append(str(cur_id))
            new_story.append({"ID": str(cur_id), "link": link})
            cur_id += 1
        else:
            recommended_stories.append(story_finder['ID'].values[0])
    
    plain_data.at[i, "rec_id"] = str(recommended_stories)
```
