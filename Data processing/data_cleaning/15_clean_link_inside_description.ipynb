```python
def remove_links(text):
    pattern_2 = r'[-a-zA-Z0-9@:%._\+~#=]{1,256}\.[a-zA-Z0-9()]{1,6}\b([-a-zA-Z0-9()@:%_\+.~#?&//=]*)'
    cleaned_text = re.sub(pattern_2, '', text)
    return cleaned_text

df["Description"] = df["Description"].apply(remove_links)
```
